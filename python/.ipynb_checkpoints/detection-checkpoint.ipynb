## Load dataset from mixture folder

```python
import functools
import json

def compare(a, b):
    return int(a.replace('.json','')) - int(b.replace('.json', ''))

data = []
names = []

for dirname, _, filenames in os.walk('../data/output/mixtures'):
    filenames = sorted(filenames, key = functools.cmp_to_key(compare))
    for filename in filenames:
        with open(os.path.join(dirname, filename )) as json_file:
            row = json.load(json_file)
            data.append(row)
            names.append(filename)    

len(data[0])
```
