We have created a synthetic dataset of complex mixtures containing up to 92 small molecules. Each sample is a random mixture of these 92 molecules. Each of them having a random concentration  x ~ N(1, 0,5). No negative concentrations are allowed, so the distributions is truncated for x < 0. 

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.naive_bayes import GaussianNB
#  importando los modulos confusion_matrix, roc_curve, auc, accuracy_score de sklearn.metrics
from sklearn.metrics import confusion_matrix, roc_curve, auc, roc_auc_score, accuracy_score, classification_report
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
from sklearn.model_selection import train_test_split
import os

pd.set_option('display.max_rows', 500)
```

## Load dataset from mixture folder

```python
import functools
import json

def compare(a, b):
    return int(a.replace('.json','')) - int(b.replace('.json', ''))

data = []
names = []

for dirname, _, filenames in os.walk('../data/output/mixtures'):
    filenames = sorted(filenames, key = functools.cmp_to_key(compare))
    for filename in filenames:
        with open(os.path.join(dirname, filename )) as json_file:
            row = json.load(json_file)
            data.append(row)
            names.append(filename)    
            
data = pd.DataFrame(data)
data = data.div(data.sum(axis = 1), axis = 0)

len(data[0])
```

# Load the metainfo

```python
with open('../data/output/concentrations.json') as json_file:
            concentrations = pd.DataFrame(json.load(json_file))
with open('../data/output/elements.json') as json_file:
            ids = json.load(json_file)
        
# Normalize the concentrations
concentrations = concentrations / concentrations.sum(axis = 0)
```

```python
figure(num=None, figsize=(18, 6), dpi=80, facecolor='w', edgecolor='k')

index = 9
part = data[2][700:1024]
print(concentrations[index])
plt.plot(list(range(0,len(part))), part)
plt.show()
```

```python
bases = []
for entry in ids:
    with open(os.path.join('../data/spectra', entry['name'])) as json_file:
        bases.append(json.load(json_file))
bases = pd.DataFrame(bases)       
bases = bases.div(bases.sum(axis = 1), axis = 0)
part = bases.loc[16, 0:1024]
figure(num=None, figsize=(18, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(list(range(0,len(part))), part)
plt.show()
```
